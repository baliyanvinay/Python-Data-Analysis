# Drop out Rate among States

```python
import pandas as pd
```

```python
df_dropout=pd.read_csv("..\Data\Indian School Education Statistics\dropout-ratio-2012-2015.csv")
```

```python
df_dropout.head()
```

```python
df_dropout['year'] = df_dropout['year'].str[:4]
df_dropout.head()
```

### Drop out rate of Uttar Pradesh through out years for Primary, Secondary and Higher Secondary

```python
df_dropout['State_UT'].unique()
```

```python
df_dropout_UP=df_dropout.loc[df_dropout['State_UT']=='Uttar Pradesh']
df_dropout_UP
```

```python
type(df_dropout_UP)
```

```python
df_dropout_UP.drop(columns={'State_UT'})
df_dropout_UP
```

```python
df_dropout_UP.set_index('year',inplace=True)
```

```python
df_dropout_UP
```

```python
import matplotlib.pyplot as plt
```

```python
df_dropout_UP.replace('NR','0', inplace=True)
df_dropout_UP
```

```python
df_dropout_UP=df_dropout_UP.astype(float)
```

```python
df_dropout_UP[['Primary_Boys','Primary_Girls','Primary_Total']].plot(kind='line',  figsize=(20,6), style='-.', grid=True)

plt.title('Drop out of Primary School in Uttar Pradesh', size=15)
plt.xlabel('Year', size=12)
plt.ylabel('Drop out rate', size=12)

plt.show()
```
